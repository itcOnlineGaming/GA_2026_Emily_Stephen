```python
import pandas as pd
df = pd.read_csv("game_results.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["player_speed"], df["score"])
plt.xlabel("Player Speed")
plt.ylabel("Score")
plt.title("Score vs Player Speed")
plt.show()
```

```python
avg_scores = df.groupby("player_speed")["score"].mean()

plt.figure()
avg_scores.plot(kind="bar")
plt.xlabel("Player Speed")
plt.ylabel("Average Score")
plt.title("Average Score by Player Speed")
plt.show()
```
