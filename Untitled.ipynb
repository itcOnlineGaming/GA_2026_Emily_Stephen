```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the combined CSV file
df = pd.read_csv("combined_data.csv")

# Create scatter plot
plt.figure()
plt.scatter(df["player_speed"], df["score"])
plt.xlabel("Player Speed")
plt.ylabel("Score")
plt.title("Score vs Player Speed")
plt.show()

# Calculate average scores by player speed
avg_scores = df.groupby("player_speed")["score"].mean()
print("\nAverage Scores by Player Speed:")
print(avg_scores)
```
